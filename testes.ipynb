```python
import pandas as pd
df = pd.read_csv("dados_saude.csv")
df
```

```python
df.plot(kind="scatter", x="idade", y="atividade", z="colesterol");
```

```python
from sklearn import linear_model
modelo = linear_model.LinearRegression()
x = df[["idade" , "atividade"]]
y = df[["colesterol"]]

modelo.fit(x, y,)
```

```python
print(modelo.predict([[50,50]])[0][0])
```
